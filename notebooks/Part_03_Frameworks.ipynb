# 1. Introduction

## 1.1 What Are Agentic Frameworks?
Key reference: Agentic frameworks such as LangGraph and CrewAI embody this dynamic approach, enabling developers to move beyond single-agent, linear workflows into multi-agent, collaborative systems​​.

## 1.2 Why Are They Important?
At their core, agent framework offers:
- A simple way to define agents and tools
- An orchestration mechanism
- State management
- Additional tools to enable more complex applications like: persistence layer (memory), interruptions, etc

# 4. ‍Deep Dive Into LangGraph

Tạo ra các ứng dụng AI đơn hoặc đa tác tử dựa trên đồ thị
## 4.1 What is a Graph?
Tưởng tượng có một loạt dữ liệu có thể biểu diễn dưới dạng mạng lưới, trong đó mỗi dữ liệu hoặc thực thể có mối quan hệ với các thực thể khác. Có hai thành phần chính tron đồ thị gồm các nốt và các cạnh.

Có hai loại đồ thị:
- Có hướng: Các cạnh biểu thị mối quan hệ giữa các nút (Vd: Theo dõi trên mạng xã hội)
- Vô hướng: Các cạnh không có hướng và biểu thị mối quan hệ đối xứng (Vd: Kết nối trên LinkedIn)



## 4.2 Key Concepts
### 4.2.1 Graph Structures
Trọng tâm thiết kế của LangGraph là biểu diễn quy trình làm việc của ứng dụng dưới dạng đồ thị. Đồ thị này bao gồm hai thành phần chính:
- Nodes: mỗi nút đại diện cho một đơn vị công việc hoặc hành động riêng biệt trong ứng dụng. Các node này về cơ bản là các hàm python đóng gói một tác vụ cụ thể, các tác vụ này có thể nhiều thao tác khác nhau (giao tiếp với LLM để tạo văn bản, tương tác với các API công cụ bên ngoài, thao tác dữ liệu thông qua quy trình, tương tác với người dùng để thu thập thông tin...)
- Edge: các cạnh hướng dẫn luồng thôn tin và điều khiển và chỉ định trình tự thao tác. có nhiều loại cạnh:
  - Simple Edge: Biểu thị một luồng trực tiếp và không điều kiện từ node này sang node khác. Đầu ra của node đầu tiền làm đầu vào của node tiếp theo, tạo ra một quy trình tuyến tính.
  - Conditional Edge: Cạnh có điều kiên cho phép quy trình làm việc phân nhánh dựa trên kết quả hoạt động của một nút cụ thể. 



### 4.2.2 State Management
Một sự chủ yếu rằng quản lý các agent thì phải đảm bảo chúng phải hoạt động với sự hiểu biết chung với trạng thái hiện tại của công việc. LangGraph giải quyết điều này thông qua quản lý trạng thái tự động. Điều này có nghĩa là thư viện sẽ xử lý tracking và update trạng thái của nodes khi thực hiện nhiệm vụ của nó.

Đối tượng hoạt động như một kho lưu trữ thông tin quan trọng cần được tại nhiều thời điểm khác nhau trong quy trình làm việc, bao gồm:
- Conversation History: Duy trì trạng thái hội thoại
- Contextual data: Dữ liệu thông tin liên quan đến một nhiệm vụ hiện tại (sở thích, hành vi trước đây, dữ liệu bên ngoài liên quan...) có thể được lưu trữ trong trạng thái để các tác nhân sử dụng khi ra quyết định.
- Internal variables: Các tác nhân có thể sử dụng trạng thái để theo dõi các cờ nội bộ, bộ đếm hoặc các biến khác hướng dẫn hành vi và quá trình ra quyết định của chúng.

## 5.2 Creating a Basic Chatbot in LangGraph
- Import Necessary Libraries
- Define the State Structure
- Initialize the LLM
- Create the Chatbot Node
- Define Entry and Finish Points
- Compile the Graph
- Visualize the Graph
- Run the Chatbot

```python
# Import Necessary Libraries
from typing import Annotated
from typing_extensions import TypedDict
from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages

# Define the State Structure
class State(TypedDict):
    # "message" stores the conversation history
    # "add_messages" function ensures new messages are appended to the list
    messages: Annotated[list, add_messages]

# Create an instance of the  StateGraph, passing the State class
graph_builder = StateGraph(State)

# Initialize the LLM
from langchain_anthropic import ChatAnthropic

api_key = "your_api_key_here"  # Replace with your actual API key
llm = ChatAnthropic(model='claude-3-5-sonnet-20240620', api_key=api_key)

# Create the Chatbot Node
def chatbot(state: State):
    # Use LLM to generate based on the current conversation history
    response = llm.invoke(state['messages'])

    # Return the updated state with the new message appended
    return {"messages": [response]}

# Add the chatbot node to the graph
graph_builder.add_node('chatbot', chatbot)

# Define Entry and Finish Points
# For this basic chatbot, the 'chatbot' node is both the entry and finish point
graph_builder.add_edge(START, 'chatbot')
graph_builder.add_edge('chatbot', END)

# Compile the Graph
graph = graph_builder.compile()

# Visualize the Graph
from IPython.display import Image, display
try:
    display(Image(graph.get_graph().draw_mermaid_png()))
except Exception:
    # This requires some extra dependencies and is optional
    pass
```

```python
# Run the chatbot
while True:
    user_input = input("User: ")
    if user_input.lower() in ['quit', 'exit', 'q']:
        print("Goodbye!")
        break

    # Process user input through LangGraph
    for event in graph.stream({'message': [("user", user_input)]}):
        for value in event.values():
            print("Assistant: ", value['messages'][-1].content)
```

## 5.3 Advanced LangGraph Techniques
Enhancing our Basic Chatbot with Tool Integration

Thêm chức năng giúp chatbot có thể search web

```python
from typing import Annotated
from langchain_anthropic import ChatAnthropic
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_core.messages import BaseMessage
from typing_extensions import TypedDict

from langgraph.graph import StateGraph
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode, tools_condition

class State(TypedDict):
    messages: Annotated[list, add_messages]

graph_builder = StateGraph(State)


tool = TavilySearchResults(max_results=2, tavily_api_key='tavily_api_key')
tools = [tool]
llm = ChatAnthropic(model="claude-3-5-sonnet-20240620", api_key=api_key)
llm_with_tools = llm.bind_tools(tools)


def chatbot(state: State):
    return {'messages': [llm_with_tools.invoke(state['messages'])]}

graph_builder.add_node('chatbot', chatbot)

tool_node = ToolNode(tools=[tool])
graph_builder.add_node('tools', tool_node)

graph_builder.add_conditional_edges(
    'chatbot', 
    tools_condition
)
# Anytime a tool is called, we return to the chatbot to decide the next step
graph_builder.add_edge('tools', 'chatbot')
graph_builder.set_entry_point('chatbot')

graph = graph_builder.compile()
display(Image(graph.get_graph().draw_mermaid_png()))
```

**Explaination:**
1. Import the Tool
2. Define and Bind the Tool: Tạo một công cụ và liên kết nó với LLM bằng bind_tools(), điều này thông báo cho LLM về các công cụ khả dụng và cách sử dụng của chúng.
3. Create a ToolNode
4. Add the ToolNode to the Graph
5. Conditional Routing: Thiết lập logic định tuyến dựa trên việc LLM có quyết định gọi công cụ hay không. Hàm tools_condition kiểm tra xem phản hồi của LLM có bao gồm các lệnh gọi công cụ hay không.
6. Loop Back: Sau khi xử lý thì điều hướng về chatbot để tiếp tục công việc

Bây giờ khi hỏi một thông tin bên ngoài, LLM có thể chọn công cụ để tìm kiếm thông tin (hoặc retrieval...) rồi đưa ra phản hồi

# 6. Adding Memory to the Chatbot

LangGraph's Checkpointing System:
- Checkpointer: Khi biên dịch LangGraph, có thể cung cấp một đối tượng checkpoint. Nó chịu trách nhiệm lưu trạng thái của graph của các thời điểm khác nhau
- Thread ID: Mỗi lần invoke vào graph, sẽ cung cấp một thread_id. ID này được dùng bởi checkpointer để track các chủ đề hội thoại khác nhau
- Automatic Saving and Loading: LangGraph tự động lưu các trạng thái sau mỗi bước thực thi của graph cho thread_id được truyền vào. Khi invoke graph lần nữa với thread_id tương tự. nó tự động load trạng thái đã lưu cho phép chatbot tiếp tục đoạn hội thoại nơi nó off.

```python
# Implementing Memory with Checkpointing
# ... (previous code to define State, graph_builder, nodes and edges)
from langgraph.checkpoint.memory import MemorySaver

# Create a MemorySaver object to act as the checkpointer
memory = MemorySaver()

# Compile the graph, passing in the 'memory object as the checkpointer
graph = graph_builder.compile(checkpointer=memory)
display(Image(graph.get_graph().draw_mermaid_png()))
# ... (Rest of the code to run the chatbbot)
```

```python
config = {'configurable': {'thread_id': '1'}}

while True:
    user_input = input("User: ")
    if user_input.lower() in ['quit', 'exit', 'q']:
        print("Goodbye!")
        break

    # Process user input through LangGraph
    for event in graph.stream({"message": [("user", user_input)]}, config):
        for value in event.values():
            print("Assustant: ", value['messages'][-1].content)
            
```

# 7. Human in the Loop

Quy trình làm việc có sự tham gia của con người là điều cần thiết trong những tình huống mà bạn muốn kết hợp sự giám sát, xác minh hoặc ra quyết định của con người vào ứng dụng AI của mình

```python
# Implementing Human-in-the-Loop with Interrupts
# Dưới đây là ví dụ áp dụng interrupt_before và interrupt_after để yêu cầu can thiệp của con người
from typing import Annotated

from langchain_anthropic import ChatAnthropic
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_core.messages import BaseMessage
from typing_extensions import TypedDict

from langgraph.checkpoint.memory import MemorySaver
from langgraph.graph import StateGraph
from langgraph.graph.message import add_messages
from langgraph.prebuilt import ToolNode, tools_condition


class State(TypedDict):
    messages: Annotated[list, add_messages]


graph_builder = StateGraph(State)


tool = TavilySearchResults(max_results=2, tavily_api_key='tavily_api_key')
tools = [tool]
llm = ChatAnthropic(model="claude-3-5-sonnet-20240620")
llm_with_tools = llm.bind_tools(tools)


def chatbot(state: State):
    return {"messages": [llm_with_tools.invoke(state["messages"])]}


graph_builder.add_node("chatbot", chatbot)

tool_node = ToolNode(tools=[tool])
graph_builder.add_node("tools", tool_node)

graph_builder.add_conditional_edges(
    "chatbot",
    tools_condition,
)
graph_builder.add_edge("tools", "chatbot")
graph_builder.set_entry_point("chatbot")

memory = MemorySaver()
graph = graph_builder.compile(
    checkpointer=memory,
    # This is new!
    interrupt_before=["tools"],
    # Note: can also interrupt __after__ actions, if desired.
    # interrupt_after=["tools"]
)
display(Image(graph.get_graph().draw_mermaid_png()))
```

**Explaination:**
Trong ví dụ này, đồ thị tạm dừng ngay trước khi thực thi công cụ. Nút công cụ này chịu bất kỳ trách nhiệm thực hiện các công việc sau đó. Bằng cách này, chúng ta đang:
- Approve the Tool Call: Cho phép con người xem xét và chấp thuận việc chạy tool
- Modify the Tool Call:Nếu con người thấy cần phải điều chỉnh lệnh gọi công cụ của LLM (vd: tinh chỉnh nội dung tìm kiếm) và sửa đổi trạng thái của graph rồi thực thi
- Modify the Tool Call: Con người có thể quyết định không chạy tool nếu không cần thiết.
