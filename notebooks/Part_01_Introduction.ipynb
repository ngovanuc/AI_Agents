# 2. What are AI agents?

Tác nhân AI là một hệ thống có thể nhận biết môi trường xung quanh thông qua các thông tin/tín hiệu đầu vào, xử lý thông tin và tác động lên môi trường thông qua các bộ truyền động (công cụ) để đạt được mục tiêu cụ thể

Hình dung một cách đơn giản chúng là một thực thể kỹ thuật có thể quan sát, suy nghĩ và hành động - tương tự như cách con người tương tác với môi trường xung quanh, nhưng theo một cách được lập trình sẵn và có mục đích



## 2.1 Characteristics of AI Agents
Các tác nhân AI sở hữu những đặc điểm chính:
- Autonomy (tính tự chủ): hoạt động mà không cần sự can thiệp của con người, tự đưa ra quyết định.
- Reactivate and Proactivate behavior (hành vi phản ứng và chủ động): phản ứng với những thay đổi của môi trường và thực hiện các biện pháp chủ động để đạt được mục tiêu
- Adaptability (khả năng thích nghi): học hỏi và phát triển bằng cách xử lý thông tin và kinh nghiệm mới
- Goal-oriented (hướng tới mục tiêu): làm việc để đạt được các mục tiêu đã xác định hoặc tối ưu hóa kết quả
- Interactivity (tính tương tác): giao tiếp và hợp tác với các tác nhân hoặc con người.
- Persistance (tính bền bỉ): hoạt động liên tục, giám sát và phản ứng với các môi trường biến động


# 3. Core Components of AI Agents

Về cơ bản, một AI agent có những thành phần như sau:
- Perception (sự nhận thức)
- Reasoning (khả năng suy luận)
- Action (hành động)
- Knowledge Base
- Learning
- Communication Interface



## 3.1 Perception (Sensors)
Những yếu tố này giúp tác nhân nhận biết môi trường xung quanh. Chúng có thể là cảm biến vật lý (micro, camera...) hoặc đầu vào kỹ thuật số (luồng dữ liệu, tương tác của người dùng...)

## 3.2 Reasoning (Processor)
Được xem như là bộ não của tác nhân, xử lý thông tin từ các cảm biến và xác định các hành động phù hợp. Thành phần này triển khai các thuật toán ra quyết định của các nhân và duy trì mọi trạng thái nội bộ cần thiết.

Các tác nhân AI sử dụng nhiều cơ chế đưa ra quyết định khác nhau

## 3.3 Action (Actuators)
Phương tiện mà một tác nhân tác động đến môi trường của nó hoặc đơn giản là cho phép tác nhân thực hiện hành động. chúng có thể là vật lý (cánh tay robot...) hoặc tín hiệu kỹ thuật số

# 3.4 Knowledge Base
Dựa trên cơ sở tri thức mà tác nhân sử dụng

## 3.5 Learning
Cho phép tác nhân cải thiện hiệu suất theo thời gian bằng cách học hỏi từ dữ liệu và kinh nghiệm. Nó sử dụng các kỹ thuật học có giám sát, không giám sát và học tăng cường...

## 3.6 Communication Interface
Một giao diện cho phép con người có thể sử dụng



# 4. How AI Agents Interact with Their Environment

Chu trình tương tác này thường được gọi là chu trình "cảm nhận - lập kế hoạch - hành đông" (hay còn gọi chu trình "nhận thức - hành động")

Lấy ví dụ xe tự hành:

## 4.1 Perception Phase
**Sensors → Processing → State Update**
- Agent nhận thông tin đầu vào thông qua sensor
- Thông tin được xử lý và giải thích
- Trạng thái hiện tại được cập nhật dựa trên thông tin mới

## 4.2 Decision Phase
**Current State + Goals → Evaluate Options → Select Best Action**
- Agent đánh giá tính khả thi của hành động
- Cân nhắc mục tiêu và hạn chế
- Lựa chọn hành động tốt nhất dựa trên thông tin có được

## 4.3 Action Phase
**Execute Action → Observe Changes → Begin New Cycle**
- Hành động được chọn được thực hiện thông qua bộ truyền động
- Môi trường thay đổi theo như kết quả
- Tác nhân giám sát kết quả thông qua cảm biến và tiếp tục chu trình mới

Chu kỳ lặp lại liên tục, thường là nhiều lần mỗi giây. Điều làm nên sức mạnh của chu kỳ này là:
- Adaptability (khả năng thích ứng): Nếu có điều gì bất ngờ xảy ra, tác nhân có thể phát hiện ra điều này trong chu kỳ tiếp theo và điều chỉnh hành vi của mình cho phù hợp.
- Learning Opportunities (cơ hội học tập): Tác nhân so sánh kết quả dự đoán so với kết quả thực tế để cải thiện hiệu suất trong tương lai
- Goal-Directed Behavior (hành vi hướng tới mục tiêu): Mỗi chu kỳ đưa tác nhân tiến tới gần hơn với mục tiêu của nó trong khi vẫn tôn trọng các ràng buộc

# 5. How AI Agents Function?

Sau đây là quy trình khi thực hiện truy vấn một AI agent
## 5.1 Orchestration layer (The Control Center)
Tưởng tượng yêu cầu AI agent lên lịch họp



Truy vấn sẽ được xử lý bởi lớp điều phối (trung tâm điều khiển của tác nhân)

Có 04 chúc năng chính của lớp điều phối:
- Memory: Duy trì trí nhớ về toàn bộ tương tác người dùng
- State: Lưu giữ trạng thái hiện tại của toàn bộ quá trình
- Reasoning: Hướng dẫn agent suy nghĩ
- Planning: Lập kế hoạch và quyết định nên làm gì tiếp theo



## 5.2 Models (The Brain)
Các mô hình là trung tâm quyết định cho toàn bộ tác nhân. Model có thể là AI model, LLM...



Để hiểu truy vấn, xây dựng kế hoạch và xác định hành động tiếp theo, mô hình sử dụng các khuôn khổ lý luận và logic:
- ReAct (Reason + Act): Đảm bảo các hành động chu đáo và có chủ đích
- Chain-of-Thought: Lí luận theo chuỗi suy nghĩ thông qua các bước trung gian
- Tree-of-Thoughts: Khám phá nhiều con đường để tìm ra giải pháp tốt nhất

=> Mô hình cần xác định những hành động và thực hiện những hành động đó bằng các công cụ cụ thể

## 5.3 Tools (The Hands)
Bằng cách sử dụng công cụ, các tác nhân có thể tương tác với thế giới bên ngoài



Các công cụ cho phép tác nhân thực hiện các hành động vượt quá khả năng của mô hình, truy cập thông tin thời gian thực hoặc hoàn thành các nhiệm vụ trong thế giới thực. 

# . ✅ When to use agents / ⛔ When to avoid them

✅ Khi nào nên dùng agents:
- Khi workflow (luồng công việc) không thể xác định trước một cách rõ ràng. Ví dụ: người dùng có yêu cầu phức tạp như "Tôi có thể đến vào thứ Hai nhưng quên hộ chiếu, có thể đến muộn đến thứ Tư, vậy có thể đi lướt sóng sáng thứ Ba kèm bảo hiểm hủy không?"
- Khi cần sự linh hoạt trong việc xử lý yêu cầu phức tạp, đa bước.
- Khi muốn LLM tự quyết định luồng xử lý phù hợp để hoàn thành nhiệm vụ.
- Khi kết hợp được nhiều công cụ/phần mềm:
- API thời tiết, Google Maps, dashboard nội bộ, hệ thống RAG...

⛔ Khi nào nên tránh dùng agents:
- Khi luồng công việc có thể xác định trước một cách rõ ràng và bao phủ mọi tình huống. Ví dụ: Người dùng chỉ muốn tìm hiểu thông tin chuyến đi → hiển thị ô tìm kiếm.
- Người dùng muốn liên hệ bán hàng → hiển thị form liên hệ.
- Khi không cần sự linh hoạt, vì:
- Có thể code tay toàn bộ logic → hệ thống ổn định, đáng tin cậy 100%.
- Giảm rủi ro do hành vi không đoán trước được từ LLM.
- Nguyên tắc gợi ý: Nếu có thể đơn giản hóa mà vẫn xử lý tốt → nên tránh dùng agent.

=> Chỉ nên dùng agents khi workflow cứng nhắc không còn đủ để xử lý tác vụ.
Nếu workflow đã rõ ràng và đầy đủ, cứ code tay cho chắc!
Sự linh hoạt là cái giá phải trả, nên chỉ dùng khi thật sự cần.

# 7. Application Areas

# 8. Conclusion

⚙️ AI Agents là gì?
- Là các hệ thống tự trị có thể nhận biết (perceive), ra quyết định (decide) và hành động (act) để đạt được mục tiêu.
- Cấu trúc lõi gồm:
    + Cảm biến (Sensors) – thu thập thông tin từ môi trường.
    + Bộ truyền động (Actuators) – thực hiện hành động.
    + Bộ ra quyết định (Decision Engine) – phân tích và lựa chọn hành động.
    + Module học (Learning Module) – học từ dữ liệu để cải thiện hiệu suất.

🌐 Ứng dụng thực tiễn của AI Agents:
- Trợ lý ảo (Virtual Assistants) – như Siri, Alexa, v.v.
- Xe tự hành (Autonomous Vehicles).
- Y tế thông minh (Healthcare) – hỗ trợ chẩn đoán, theo dõi bệnh nhân, cá nhân hóa điều trị.

⚠️ Thách thức trong việc xây dựng AI Agents:
- Vấn đề đạo đức (Ethical concerns) – quyền riêng tư, thiên vị, trách nhiệm.
- Phụ thuộc dữ liệu (Data Dependency) – yêu cầu dữ liệu lớn và chất lượng cao.
- Khả năng mở rộng (Scalability) – làm sao để agent hoạt động hiệu quả khi tăng quy mô hoặc độ phức tạp.

🔮 Tương lai của AI Agents:
- Hướng tới AI tổng quát (General AI) – linh hoạt, đa nhiệm như con người.
- Tăng cường hợp tác người–máy (Human-Agent Collaboration) – phối hợp thông minh và hiệu quả.
- Tập trung vào giá trị con người (Human-aligned values) – đảm bảo AI phục vụ xã hội, không gây hại.
